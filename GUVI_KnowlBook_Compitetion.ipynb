# **Multiple Plots**

'matplotlib' allows you to create multiple plots within the same figure, either in a single row or column or using a grid layout. This feature is useful when you want to visualize multiple related plots side by side for comparison or to display different aspects of your data.

## ***1. Single Row Layout***

To create plots side by side in a single row, you can use the plt.subplot() function. The function takes three arguments: the number of rows, the number of columns, and the plot number.



```python
import matplotlib.pyplot as plt

# Data for first plot
x1 = [1, 2, 3, 4, 5]
y1 = [10, 25, 18, 30, 15]

# Data for second plot
x2 = [1, 2, 3, 4, 5]
y2 = [5, 12, 8, 15, 7]

# Create the first plot
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('Plot 1')

# Create the second plot
plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.title('Plot 2')

plt.tight_layout()  # Ensures plots don't overlap
plt.show()
```

# ***2. Single Column Layout***

To create plots on top of each other in a single column, you can use the same plt.subplot() function.

```python
import matplotlib.pyplot as plt

# Data for first plot
x1 = [1, 2, 3, 4, 5]
y1 = [10, 25, 18, 30, 15]

# Data for second plot
x2 = [1, 2, 3, 4, 5]
y2 = [5, 12, 8, 15, 7]

# Create the first plot
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('Plot 1')

# Create the second plot
plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('Plot 2')

plt.tight_layout()  # Ensures plots don't overlap
plt.show()
```

# ***3. Grid Layout***

To create a grid of plots, you can use plt.subplots().

```python
import matplotlib.pyplot as plt

# Data for first plot
x1 = [1, 2, 3, 4, 5]
y1 = [10, 25, 18, 30, 15]

# Data for second plot
x2 = [1, 2, 3, 4, 5]
y2 = [5, 12, 8, 15, 7]

# Create a 1x2 grid of plots
fig, axs = plt.subplots(1, 2)

# Plot the first data on the first plot
axs[0].plot(x1, y1)
axs[0].set_title('Plot 1')

# Plot the second data on the second plot
axs[1].plot(x2, y2)
axs[1].set_title('Plot 2')

plt.tight_layout()  # Ensures plots don't overlap
plt.show()
```

# **Matplotlib - Line Graph**

A line graph, also known as a line plot or line chart, is a type of data visualization used to display data points connected by straight lines. It is commonly used to visualize trends, changes, or relationships over time or a continuous interval. Line graphs are particularly useful for showing how one variable changes in response to another variable or as an independent variable evolves.

**Components of a Line Plots:**

1. **Data Points:** Each data point represents a value along the x-axis (horizontal axis) and its corresponding value on the y-axis (vertical axis). These points are connected by straight lines to form the graph.

2. **X-Axis:** The horizontal axis, also known as the x-axis, typically represents the independent variable, which is the variable that is being measured or controlled in the study. It may represent time, continuous values, or categories.

3. **Y-Axis:** The vertical axis, also known as the y-axis, represents the dependent variable, which is the variable being observed or measured in response to changes in the independent variable.

4. **Lines:** The lines connecting the data points help to visualize the trend or pattern in the data. The slope of the line provides information about the direction and magnitude of the change.

5. **Title:** The graph should have a descriptive title that summarizes what is being visualized.

6. **Axis Labels:** Labels on the x-axis and y-axis should provide context and describe the data being displayed.

7. **Legend:** A legend is used when there are multiple lines on the graph, each representing a different dataset or category. It helps identify which line corresponds to which data.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperatures = [15, 18, 20, 25, 28, 30, 32, 31, 28, 25, 20, 17]

plt.plot(months, temperatures, marker='o', color='blue')
plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.title('Average Monthly Temperature')
plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, the line graph will show the trend of how the average monthly temperature changes over the course of a year in the city. The points are connected by lines, giving a clear visual representation of temperature fluctuations throughout the year.

# **Line Graph Customization Options:**

1. **Line Style**: You can customize the style of the line in the line graph using the `linestyle` parameter. Common line styles include '-', '--', '-.', and ':' for solid, dashed, dash-dot, and dotted lines, respectively.

2. **Line Color**: The `color` parameter allows you to set the color of the line in the graph. You can use named colors (e.g., 'red', 'blue') or specify colors using hexadecimal codes (e.g., '#FF0000' for red).

3. **Line Width**: The `linewidth` or `lw` parameter lets you adjust the thickness of the line in the graph.

4. **Marker Style**: You can add markers at each data point using the `marker` parameter. Common markers include 'o' for circles, 's' for squares, 'x' for crosses, and '^' for triangles.

5. **Marker Size**: The `markersize` parameter allows you to set the size of the markers in the graph.

6. **Marker Color**: The `markerfacecolor` or `mfc` parameter can be used to set the color of the markers.

7. **Marker Edge Color**: The `markeredgecolor` or `mec` parameter sets the color of the marker edges.

8. **Labeling Data Points**: You can label specific data points in the line graph using the `label` parameter. This is helpful when you want to add a legend to the graph.

9. **Legend**: To create a legend for the line graph, use the `plt.legend()` function. The labels provided in the `label` parameter of the `plot()` function will be used as legend entries.

**Common Use Cases of Line Graphs:**

1. **Time Series Data**: Line graphs are commonly used to visualize time series data, such as stock prices, temperature trends, or website traffic over time.

2. **Trend Analysis**: Line graphs help identify trends, fluctuations, and patterns in data. They are valuable in understanding changes in variables over time or across different conditions.

3. **Comparing Multiple Variables**: Line graphs are useful for comparing multiple variables or datasets on the same graph. By plotting multiple lines, you can observe how different variables evolve over the same time period.

4. **Forecasting**: Line graphs can aid in forecasting future trends based on historical data patterns.

5. **Performance Metrics**: Line graphs are widely used to visualize performance metrics, such as sales revenue, profit margins, or user engagement metrics.

6. **Process Improvement**: Line graphs can show the performance of a process over time, helping identify opportunities for improvement.

7. **Experimental Data**: In scientific research, line graphs are used to visualize experimental data and relationships between variables.

Remember, when customizing line graphs, it's essential to choose colors, line styles, and markers that enhance readability and clarity of the data. Additionally, adding labels, legends, and appropriate axis titles helps convey the information effectively.

# **Matplotlib - Charts**

Charts, also known as graphs or visualizations, are graphical representations of data and information. They are used to present data in a visual and easily understandable format, making it easier to identify patterns, trends, and insights within the data. Charts are an essential tool for data analysis, communication, and decision-making in various fields, including business, finance, science, and research.

Here are some common types of charts:

1. **Line Chart**: Shows data points connected by straight lines, commonly used to display trends over time or continuous intervals.

2. **Bar Chart**: Represents data with rectangular bars, suitable for comparing discrete categories or values.

3. **Pie Chart**: Displays data as slices of a circle, ideal for showing proportions or percentages of a whole.

4. **Scatter Plot**: Plots individual data points as dots, useful for visualizing the relationship between two variables.

5. **Histogram**: Illustrates the distribution of continuous data by grouping it into bins.

6. **Area Chart**: Similar to a line chart, but the area between the line and the x-axis is filled, making it useful for showing cumulative data.

7. **Stacked Bar Chart**: A variation of the bar chart where the bars are stacked to show the composition of different categories.

8. **Box Plot (Box-and-Whisker Plot)**: Represents the distribution of data through quartiles, making it useful for visualizing the spread of data.

9. **Heatmap**: A two-dimensional representation of data using colors, useful for showing correlations and patterns in large datasets.

10. **Radar Chart (Spider Chart)**: Displays multivariate data on a two-dimensional chart, with each variable represented as a spoke on the radar.

11. **Bubble Chart**: A variation of the scatter plot where the size of the marker represents a third dimension of the data.

12. **Gantt Chart**: A type of bar chart used in project management to show the schedule and progress of tasks over time.

13. **Sankey Diagram**: Illustrates the flow of data or resources from one category to another.

14. **Word Cloud**: Represents textual data by visualizing frequently occurring words with varying sizes based on their frequency.

15. **Tree Map**: Displays hierarchical data using nested rectangles, with sizes corresponding to the values.

16. **Network Graph**: Shows relationships between entities as nodes connected by edges.

The choice of the chart depends on the type of data you have and the specific insights you want to convey. Selecting the right chart is crucial for effective data communication and analysis. Different visualization libraries in Python, such as Matplotlib, Seaborn, Plotly, and Altair, offer support for creating these various types of charts. Let see only one chart in the above list.


##**Bar Chart**

A bar chart is a type of chart that uses rectangular bars to represent data. It is used to compare different categories or discrete data points and visualize their values. The length or height of the bars is proportional to the data being represented, making it easy to compare the magnitudes of different items. Bar charts are effective in showing data distributions, identifying trends, and making comparisons between different groups.

There are two main types of bar charts:

1. **Vertical Bar Chart:** In a vertical bar chart, the bars are plotted vertically along the y-axis (vertical axis), and the categories or data points are displayed along the x-axis (horizontal axis).

2. **Horizontal Bar Chart:** In a horizontal bar chart, the bars are plotted horizontally along the x-axis, and the categories or data points are displayed along the y-axis.

When to Use Bar Charts:

- Comparing categorical data or discrete items.
- Showing the distribution of data across different categories.
- Visualizing data with distinct categories or groups.
- Displaying data with nominal or ordinal scales.


##**Vertical Bar Chart:**

Let's consider a simple example of a vertical bar chart showing the sales performance of different products:

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1200, 900, 1500, 800]

plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Performance')
plt.show()
```

In this example, the vertical bar chart will show four bars, each representing the sales performance of a product. The x-axis will display the product names, and the y-axis will represent the sales values. The lengths of the bars will be proportional to the sales figures, allowing easy comparison between the products.

#**Horizontal Bar Chart:**

Using the same data, let's create a horizontal bar chart to visualize the product sales:

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1200, 900, 1500, 800]

plt.barh(products, sales)
plt.xlabel('Sales')
plt.ylabel('Products')
plt.title('Product Sales Performance')
plt.show()
```

In this example, the horizontal bar chart will display the product names along the y-axis and the sales values along the x-axis. The lengths of the bars will still be proportional to the sales figures, allowing easy comparison as well.

# **Matplotlib - Histogram**

A histogram is a graphical representation of the distribution of numerical data. It divides the data into bins or intervals and displays the frequency or count of data points falling within each bin. Histograms are particularly useful for understanding the shape, spread, and central tendency of a dataset. They help identify patterns, outliers, and gaps in the data.

In a histogram, the x-axis represents the data range or bins, and the y-axis represents the frequency or count of data points within each bin.

Creating a Histogram in Matplotlib:
To create a histogram in Matplotlib, you can use the plt.hist() function. The function takes the data array and optional parameters to customize the appearance of the histogram.

```python
# Sample data (list of values)
data = [15, 20, 25, 30, 35, 30, 25, 20, 15, 10, 5, 10, 15]

# Create the histogram
plt.hist(data, bins=5, edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

plt.show()
```

In this example, the histogram is created using the data list with five bins specified by the bins parameter. The edgecolor parameter is set to 'black' to add black borders to each bar in the histogram.

Check some sample histograms:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for the histogram
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Create the histogram with custom options
plt.hist(data, bins=15, edgecolor='black', color='skyblue', alpha=0.7)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Customized Histogram')
plt.grid(True)  # Optional: Add gridlines
plt.show()
```

```python
# Generate random data for the histogram
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Create the histogram with cumulative density
plt.hist(data, bins=15, edgecolor='black', color='orange', cumulative=True, density=True)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Cumulative Density')
plt.title('Histogram with Cumulative Density')
plt.grid(True)  # Optional: Add gridlines
plt.show()
```

```python
# Generate random data for the histogram
np.random.seed(42)
data1 = np.random.normal(loc=30, scale=5, size=100)
data2 = np.random.normal(loc=50, scale=10, size=100)

# Create the stacked histogram
plt.hist([data1, data2], bins=15, edgecolor='black', label=['Data 1', 'Data 2'], alpha=0.7, stacked=True)

# Add labels, title, and legend
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
plt.legend()
plt.grid(True)  # Optional: Add gridlines
plt.show()
```


**Customization Options:**

1. **bins:** Specifies the number of bins or the bin edges for the histogram.
range: Defines the range of data values to be included in the histogram.
density: If True, the histogram will be normalized to form a probability density.

2. **color:** Sets the color of the bars in the histogram.

3. **edgecolor:** Sets the color of the edges of the bars.

4. **alpha:** Sets the transparency of the bars.

5. **rwidth:** Sets the width of the bars as a fraction of the bin width.

### **Common Use Cases:**

Analyzing data distributions and identifying data patterns.
Visualizing the spread and central tendency of continuous data.
Detecting outliers and gaps in the data.

Histograms are a fundamental tool in data analysis and are commonly used in various fields such as statistics, finance, scientific research, and data exploration. They provide valuable insights into the underlying structure and characteristics of the data.

# **Scatter Plot**

A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional coordinate system. It is used to visualize the relationship between two continuous variables and helps identify patterns, correlations, or trends in the data. Each dot in the scatter plot represents a single data point, where one variable is plotted on the x-axis and the other variable on the y-axis.


## **Creating a Scatter Plot in Matplotlib:**

To create a scatter plot in Matplotlib, you can use the plt.scatter() function. The function takes the data arrays for the x-axis and y-axis and optional parameters to customize the appearance of the scatter plot.

```python

# Sample data for the x-axis and y-axis
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 30, 15]

# Create the scatter plot
plt.scatter(x, y, marker='o', color='blue', label='Data Points')

# Add labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter Plot Example')

# Add a legend (optional)
plt.legend()

plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, the scatter plot will display five data points, where the values of x are plotted on the x-axis and the values of y on the y-axis. The dots representing the data points will be marked with circles ('o') and colored blue.

# **Pie chart**

A pie chart is a circular chart that displays data as slices of a pie, where each slice represents a portion of the whole. Pie charts are commonly used to visualize the composition of a categorical variable as a percentage of the total. Each slice's size is proportional to the quantity it represents, and the total sum of all the slices equals 100%.


## **Creating a Pie Chart in Matplotlib:**

To create a pie chart in Matplotlib, you can use the plt.pie() function. The function takes the data array and optional parameters to customize the appearance of the pie chart.


```python

# Sample data for the pie chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [30, 25, 15, 30]

# Create the pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)

# Add title
plt.title('Pie Chart Example')

plt.axis('equal')  # Equal aspect ratio ensures the pie chart appears circular.

plt.show()
```

### **Area Plot (Stacked)**

The Area Plot (Stacked) is a type of data visualization used to display the contribution of different categories to the whole, while also showing the overall trend over time or across a continuous variable. It is a variation of the regular area plot where multiple data series are stacked on top of each other to represent their cumulative contribution to the total. Each data series fills the area between its curve and the curve below it.

Area plots are particularly useful when you want to compare the parts of a whole, understand the composition of a dataset, and observe the total trend. They are commonly used to represent time series data, where each series represents a category or group, and the total area of all the categories at any point in time represents the overall value of the variable being measured.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the x-axis and y-axis
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/10)

# Create the stacked area plot
plt.fill_between(x, y1, color='blue', alpha=0.5, label='Sin(x)')
plt.fill_between(x, y1, y1+y2, color='orange', alpha=0.5, label='Cos(x)')
plt.fill_between(x, y1+y2, y1+y2+y3, color='green', alpha=0.5, label='Exp(-x/10)')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Area Plot')
plt.legend()

plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, we have three data series y1, y2, and y3, representing the functions sin(x), cos(x), and exp(-x/10), respectively. These three series are stacked on top of each other to create the stacked area plot. The fill_between function is used to fill the area between the curves of each series. The alpha parameter controls the transparency of the fill, allowing the viewer to see the overlapping areas clearly.

Stacked area plots are effective in displaying the cumulative distribution of data and can be a powerful tool for visualizing multi-dimensional datasets. They help in understanding the relative contributions of individual components to the overall trend, making it easier to identify patterns and trends in complex datasets.

### **3D Plot (Surface Plot)**

A 3D Plot, specifically a Surface Plot, is a type of data visualization used to represent three-dimensional data in a three-dimensional space. Surface plots are ideal for visualizing functions of two variables, where the value of the third variable is represented by the height or color of the surface above the x-y plane. These plots are particularly useful for understanding the relationships between two independent variables and their effect on a dependent variable.

In a Surface Plot, the x and y axes represent the two independent variables, and the z-axis represents the dependent variable. The surface of the plot is created by connecting data points from the x-y plane to their corresponding z-values. The surface can be displayed as a wireframe or shaded to indicate the variations in the z-values.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data for the x-axis and y-axis
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Sample function for the z-axis (dependent variable)
Z = np.sin(np.sqrt(X**2 + Y**2)) / (np.sqrt(X**2 + Y**2))

# Create the 3D Surface Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

plt.show()
```

In this example, we use NumPy to generate sample data for the x and y axes, and then create a sample function Z that depends on both X and Y. The plot_surface function from mpl_toolkits.mplot3d is used to create the 3D surface plot. The cmap parameter is used to specify the colormap, which determines the colors used to represent the surface.

3D Surface Plots are useful when dealing with continuous data and can help visualize complex mathematical functions, elevation maps, and physical simulations. They provide an intuitive way to explore the relationships between multiple variables and identify patterns or trends in the data.

### **Box Plot**

A Box Plot, also known as a Box-and-Whisker Plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It is an effective tool for visualizing the spread, skewness, and potential outliers in a dataset.

The Box Plot consists of a rectangular box and two whiskers. The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The whiskers extend from the box to the minimum and maximum values within a certain range, usually defined as 1.5 times the IQR. Data points beyond the whiskers are considered outliers and are plotted as individual points.

The median is represented by a horizontal line inside the box. The box itself spans from Q1 to Q3, capturing the middle 50% of the data. The length of the box represents the spread or dispersion of the data in this range.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the box plot
data = [np.random.normal(0, 1, 100),
        np.random.normal(2, 1, 100),
        np.random.normal(-2, 1, 100)]

# Create the box plot
plt.boxplot(data)

# Add labels and title
plt.xlabel('Data Series')
plt.ylabel('Values')
plt.title('Box Plot')

plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, we have three sets of data represented by the data list. The boxplot function from Matplotlib is used to create the Box Plot for each dataset. The three boxes represent the distributions of the three datasets, and the whiskers extend to show the range of data points within 1.5 times the IQR.

Box Plots are commonly used to compare the distribution of multiple datasets or to identify potential outliers. They are especially useful when dealing with skewed data or when you want to get a quick visual summary of the spread and central tendency of the data. Box Plots are widely used in statistical analysis, data exploration, and data presentation.

### **Violin Plot**

A Violin Plot is a type of data visualization that combines aspects of a box plot and a kernel density plot. It displays the distribution of data across multiple categories, allowing you to compare the shape and spread of the data for each category. Violin plots are especially useful when dealing with datasets that have multimodal distributions or unevenly distributed data points.

In a Violin Plot, each category is represented by a "violin" shape. The width of the violin at any given point represents the density of data points at that value. The thicker portions of the violin indicate regions with more data points, while the narrower portions represent regions with fewer data points. The white dot inside each violin represents the median value of the data.

On either side of the violin, there are mirrored "hinges" or "whiskers" that resemble box plot whiskers. These whiskers extend to the maximum and minimum data values, or to a certain percentage of the data distribution, depending on the plot settings.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the violin plot
data = [np.random.normal(0, 1, 100),
        np.random.normal(2, 1, 100),
        np.random.normal(-2, 1, 100)]

# Create the violin plot
plt.violinplot(data, showmedians=True)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Violin Plot')

plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, we have three sets of data represented by the data list. The violinplot function from Matplotlib is used to create the Violin Plot for each dataset. The Violin Plots show the distribution of data within each category, including the medians represented by the white dots.

Violin Plots are excellent for visualizing the distribution of data across multiple categories and identifying patterns or differences between datasets. They are commonly used in exploratory data analysis, statistical comparisons, and conveying complex data distributions in a concise manner.

In a Hexbin Plot, the x-y plane is divided into a grid of hexagons, and the number of data points falling within each hexagon is counted. ### **Polar Plot**

A Polar Plot, also known as a Polar Coordinate Plot, is a data visualization that represents data in a circular format using polar coordinates. Instead of using the traditional Cartesian coordinates (x, y), polar plots use the radial distance (r) from the origin and the angle (θ) from a reference axis (usually the positive x-axis) to plot data points. Polar plots are particularly useful when dealing with data that has angular or directional characteristics.

In a Polar Plot, data points are represented by their radial distance from the center of the plot (the origin) and their corresponding angle with respect to the reference axis. The radial distance corresponds to the magnitude or value of the data, while the angle represents the direction or phase of the data.

Polar plots are commonly used to visualize cyclic or periodic data, such as angles, compass directions, circular distributions, and patterns that repeat over time. They are often used in fields such as physics, engineering, meteorology, and astronomy.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the polar plot
theta = np.linspace(0, 2 * np.pi, 100)
r = 3 + 2 * np.cos(6 * theta)

# Create the polar plot
plt.polar(theta, r, color='green')

# Add title
plt.title('Polar Plot')

plt.show()
```

In this example, we have defined the polar coordinates theta as an array of angles ranging from 0 to 2π and r as an array representing the radial distance. The polar function from Matplotlib is used to create the Polar Plot with the specified data.

Polar Plots are helpful when you want to represent data with a cyclic or periodic nature and visualize it in a circular layout. They provide an intuitive way to display patterns and trends that repeat over time or have a directional component.

### **Contour Plot**

A Contour Plot, also known as a Contour Map or Level Plot, is a two-dimensional data visualization that represents the variation of a scalar function over a continuous domain. It is particularly useful for visualizing data that is a function of two independent variables, such as height, temperature, or pressure over a geographic area or physical space.

In a Contour Plot, the x-y plane is divided into a grid, and each grid point is assigned a value based on the scalar function. Contour lines are then drawn to connect points with the same value, creating smooth curves that represent regions of constant data values. These curves are called "contours," and they provide a visual representation of the level sets or isolines of the scalar function.

Contour Plots are useful for identifying patterns, trends, and relationships between two variables. They allow you to observe regions of high or low values, locate peaks and valleys, and detect changes in the scalar function across the domain.

### **Ternary Plot**

A Ternary Plot, also known as a Triangle Plot or Gibbs Triangle, is a data visualization that represents the composition of a three-component mixture. It is particularly useful when dealing with data that is constrained to three variables, and each variable represents a percentage or proportion of a whole.

In a Ternary Plot, the three axes form an equilateral triangle, with each corner representing one of the three components. The three sides of the triangle represent all possible combinations of the three components, and the interior of the triangle represents the feasible region where the sum of the three components is equal to 100%. Data points are plotted within the triangle, and the position of each point indicates the relative proportion of the three components in the mixture.

Ternary Plots are commonly used in various fields, including chemistry, geology, ecology, and materials science, to represent the composition of mixtures and explore relationships between the components.


```python
import matplotlib.pyplot as plt

# Sample data for the ternary plot
# The sum of the three components (A, B, C) in each data point is equal to 1.
A = [0.2, 0.6, 0.1, 0.3, 0.4]
B = [0.3, 0.1, 0.5, 0.2, 0.2]
C = [0.5, 0.3, 0.4, 0.5, 0.4]

# Create the ternary plot
plt.figure(figsize=(6, 6))
plt.scatter(A, B, c=C, cmap='viridis', s=100)

# Set axis limits and labels
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('Component A')
plt.ylabel('Component B')

# Add gridlines (optional)
plt.grid(True, linestyle='dotted')

# Add a colorbar to represent Component C
cbar = plt.colorbar()
cbar.set_label('Component C')

plt.title('Ternary Plot')
plt.show()
```

In this example, we have three sets of data: A, B, and C, representing the proportions of three components in each data point. The scatter function from Matplotlib is used to create the Ternary Plot, where A and B are the coordinates of the data points within the triangle, and C is used as a color map to represent the value of Component C.

Ternary Plots are useful for visualizing the composition of mixtures and understanding the relationships between the components. They allow you to identify patterns and trends in the data and make it easier to compare multiple mixtures in a single plot.

### **Streamplot**

A Streamplot is a data visualization used to represent the flow of a vector field or fluid flow. It is particularly useful in fluid dynamics, electromagnetism, and other fields where the direction and intensity of a vector field need to be visualized.

In a Streamplot, small arrows are drawn to show the direction of the vector field at different points in the plot. The arrows are typically placed at a regular grid of points, and their length and orientation represent the magnitude and direction of the vector at each point.

Streamplots are useful for understanding the behavior of vector fields and identifying patterns such as stagnation points, vortexes, and flow patterns. They provide an intuitive way to visualize the flow of fluid or the direction of forces in a vector field.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the streamplot
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
U = -Y
V = X

# Create the streamplot
plt.streamplot(X, Y, U, V, color='blue')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Streamplot')

plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, we have two sets of data X and Y, which represent the coordinates of the grid points, and two vector fields U and V, which represent the x and y components of the vector field, respectively. Streamplots are helpful for visualizing vector fields in a 2D space, and they are commonly used in fluid mechanics, meteorology, and engineering to study the behavior of fluid flow, magnetic fields, or other vector fields. ### **Stackplot**

In a Stackplot, each dataset is represented by an area, and the total area of all datasets at any given point represents the whole. The bottom of each area starts at the baseline (usually y=0), and the top of each area is determined by the value of the dataset at that point. The areas are stacked on top of each other to create the Stackplot.

Stackplots are particularly useful when you want to visualize the part-to-whole relationships of multiple categories over time or a continuous variable. They help to highlight the total and relative contributions of each component and how the composition changes over time or along the x-axis.

In an Errorbar Plot, each data point is represented by a marker, and error bars are drawn extending from each data point to indicate the range of uncertainty. The length of the error bars represents the size of the error or uncertainty in the data.

Errorbar Plots are commonly used to visualize experimental or measurement data and display the precision or confidence level of each data point. They provide important information about the reliability of the data and allow viewers to understand the spread or variability around each data point.



```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the errorbar plot
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 4, 6, 5])
errors = np.array([0.5, 0.8, 0.6, 0.7, 0.9])

# Create the errorbar plot
plt.errorbar(x, y, yerr=errors, fmt='o', color='blue', ecolor='red', capsize=5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Errorbar Plot')

plt.grid(True)  # Optional: Add gridlines
plt.show()
```

In this example, we have two sets of data x and y, representing the x and y coordinates of the data points, and errors, representing the magnitude of the errors associated with each data point. The errorbar function from Matplotlib is used to create the Errorbar Plot. The yerr parameter specifies the size of the errors, fmt='o' sets the marker style to circles, color='blue' sets the color of the markers, and ecolor='red' sets the color of the error bars. The capsize parameter determines the length of the caps at the ends of the error bars.

Errorbar Plots are useful for providing a visual representation of the uncertainty or variability in the data and are commonly used in scientific research, engineering, and other fields where measurements and uncertainties play a crucial role.

# **Seaborn - Different Plots**

Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for creating attractive and informative statistical graphics. It is built on top of Matplotlib and provides additional functionality and ease of use for creating various types of plots. Seaborn is particularly useful for visualizing complex datasets and statistical relationships.

Here are some of the different types of plots that you can create using Seaborn:

1. **Scatter Plot**:
   Seaborn's `scatterplot` function allows you to create scatter plots to visualize the relationship between two numerical variables. It can also incorporate categorical variables using color and/or markers.

2. **Line Plot**:
   Seaborn's `lineplot` function is used to create line plots to show the trend of numerical data over an axis (usually time).

3. **Bar Plot**:
   Seaborn's `barplot` function is used to create bar plots to compare categorical variables and display aggregate data.

4. **Count Plot**:
   Seaborn's `countplot` function is used to create count plots to show the number of occurrences of each category in a categorical variable.

5. **Box Plot**:
   Seaborn's `boxplot` function is used to create box plots to visualize the distribution of numerical data across different categories.

6. **Violin Plot**:
   Seaborn's `violinplot` function creates violin plots, which are a combination of box plots and kernel density plots, to visualize the distribution of numerical data across different categories.

7. **Histogram**:
   Seaborn's `histplot` function creates histograms to visualize the distribution of numerical data.

8. **Heatmap**:
   Seaborn's `heatmap` function creates a heatmap to visualize the correlation between variables using color intensity.

9. **Pair Plot**:
   Seaborn's `pairplot` function creates a matrix of scatter plots to visualize the relationships between multiple numerical variables in a dataset.

10. **Joint Plot**:
    Seaborn's `jointplot` function creates a scatter plot with histograms on the axes to visualize the relationship between two numerical variables.

11. **Facet Grid**:
    Seaborn's `FacetGrid` class allows you to create multiple plots, arranged in a grid, to visualize the relationships between multiple variables.

These are just some of the many types of plots that Seaborn offers. Each plot is highly customizable and can be adjusted to suit your specific data visualization needs. Seaborn also provides various color palettes and themes to make your plots visually appealing.

## **Scatter Plot (Seaborn):**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for scatter plot
tips = sns.load_dataset('tips')

# Create scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
plt.title('Scatter Plot')
plt.show()
```

# **Line Plot (Seaborn):**

```python
# Sample data for line plot
dots = sns.load_dataset('dots')

# Create line plot
sns.lineplot(data=dots, x='time', y='firing_rate', hue='align', style='choice')
plt.title('Line Plot')
plt.show()
```

# **Bar Plot (Seaborn):**

```python
# Sample data for bar plot
titanic = sns.load_dataset('titanic')

# Create bar plot
sns.barplot(data=titanic, x='class', y='survived', hue='sex')
plt.title('Bar Plot')
plt.show()
```

# **Count Plot (Seaborn):**

```python
# Create count plot
sns.countplot(data=titanic, x='class', hue='sex')
plt.title('Count Plot')
plt.show()
```

# **Box Plot (Seaborn):**

```python
# Create box plot
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('Box Plot')
plt.show()
```

# **Violin Plot (Seaborn):**

```python
# Create violin plot
sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker', split=True)
plt.title('Violin Plot')
plt.show()
```

# **Histogram (Seaborn):**

```python
# Create histogram
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Histogram')
plt.show()
```

# **Heatmap (Seaborn):**

```python
# Sample data for heatmap
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot('month', 'year', 'passengers')

# Create heatmap
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap')
plt.show()
```

# **Pair Plot (Seaborn):**

```python
# Create pair plot
sns.pairplot(data=tips, hue='sex', markers=['o', 's'], diag_kind='kde')
plt.title('Pair Plot')
plt.show()
```

# **Joint Plot (Seaborn):**

```python
# Create joint plot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
plt.title('Joint Plot')
plt.show()
```

# **Facet Grid (Seaborn):**

```python
# Create facet grid
g = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.show()
```

# **Swarm Plot:**

```python
# Create swarm plot
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Swarm Plot')
plt.show()
```

# **Strip Plot:**

```python
# Create strip plot
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', jitter=True, dodge=True)
plt.title('Strip Plot')
plt.show()
```

# **Point Plot:**

```python
# Create point plot
sns.pointplot(data=tips, x='day', y='total_bill', hue='sex', ci='sd', capsize=0.1)
plt.title('Point Plot')
plt.show()
```

# **Regression Plot (lmplot):**

```python
# Create regression plot (lmplot)
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('Regression Plot (lmplot)')
plt.show()
```

# **KDE Plot:**

```python
# Create KDE plot
sns.kdeplot(data=tips['total_bill'], shade=True)
plt.title('KDE Plot')
plt.show()
```

# **Pair Grid:**

```python
# Create pair grid
g = sns.PairGrid(tips, hue='sex')
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot, levels=4, color='b')
g.add_legend()
plt.title('Pair Grid')
plt.show()
```

# **Categorical Plot:**

```python
# Create categorical plot
sns.catplot(data=titanic, x='sex', y='survived', hue='class', kind='bar', ci=None)
plt.title('Categorical Plot')
plt.show()
```

# **Rug Plot:**

```python
# Create rug plot
sns.rugplot(data=tips['total_bill'], height=0.2)
plt.title('Rug Plot')
plt.show()
```

# **Matrix Plot (Clustermap):**

```python
# Create matrix plot (clustermap)
sns.clustermap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Matrix Plot (Clustermap)')
plt.show()
```
